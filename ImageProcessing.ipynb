```python
# Function for insert image when clik the button
window.filename = ""
def InsertImage():
    # Load image from button using OpenCV
    window.filename =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("all files","*.*"), ("jpeg files","*.jpg")))
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
# Function to find dominant color 
def findDominantRGBColor():
    # Load selected image
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    img_arr = np.asarray(img)
    
    #get RGB array
    R = img_arr[:, :, 0]
    G = img_arr[:, :, 1]
    B = img_arr[:, :, 2]
    
    #sum all binary digit from each color
    sumR = []
    sumG = []
    sumB = []
    for i in R:
        sumR.append(sum(i))
    for i in G:
        sumG.append(sum(i))
    for i in B:
        sumB.append(sum(i))
    
    #return all sum of binary digit of each color
    return sum(sumR),sum(sumG),sum(sumB)

# Grayscale The Image
def grayscale():
     # Load selected image
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    img_arr = np.asarray(img)
    
    # Get RGB array
    R = img_arr[:, :, 0]
    G = img_arr[:, :, 1]
    B = img_arr[:, :, 2]
    
    # Find most dominant color in RGB
    Rv,Gv,Bv = findDominantRGBColor()
    # If Red is dominant use R:0.6, G:0.2, B:0.2 Ratio for Grayscale
    if ((Rv > Gv) & (Rv > Bv)):
        arr_gray = (R.astype(int)*0.6) + (G.astype(int)*0.2) + (B.astype(int)*0.2)
    # If Green is dominant use R:0.2, G:0.6, B:0.2 Ratio for Grayscale
    elif ((Gv > Rv) & (Gv > Bv)):
        arr_gray = (R.astype(int)*0.2) + (G.astype(int)*0.6) + (B.astype(int)*0.2)
    # If Blue is dominant use R:0.2, G:0.2, B:0.6 Ratio for Grayscale
    else:
        arr_gray = (R.astype(int)*0.2) + (G.astype(int)*0.2) + (B.astype(int)*0.6)
    arr_gray.astype("uint8")#turn type back
    
    #create new image from Gray array
    from PIL import Image
    im = Image.fromarray(arr_gray)
    im = im.convert("RGB")
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = im)
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
# Change Brightness of image
def changeBrightness(value):
    # Load selected image
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    img_arr = np.asarray(img, dtype=np.int)
    
    # new array with new brightness from value
    from PIL import Image
    new_arr = img_arr + value
    new_arr = np.clip (new_arr,0,255)
    new_arr = np.uint8(new_arr)
  
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
# Zoomin or Zoom out
def zoom():
     # Load selected image
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    img_arr = np.asarray(img)
    new_size = ((img_arr.shape[0])*2, (img_arr.shape[1]*2), img_arr.shape[2])
    new_arr = np.full(new_size, 255)
    
    # Get RGB array
    R = img_arr[:, :, 0]
    G = img_arr[:, :, 1]
    B = img_arr[:, :, 2]
    
    new_R = []
    new_G = []
    new_B = []

    for row in range(len(R)):
        temp_R = []
        temp_G = []
        temp_B = []
        for i in R[row]:
            temp_R.extend([i, i])
        for j in G[row]:
            temp_G.extend([j, j])
        for k in B[row]:
            temp_B.extend([k, k])
        for _ in (0, 1):
            new_R.append(temp_R)
            new_G.append(temp_G)
            new_B.append(temp_B)

    for i in range(len(new_arr)):
        for j in range(len(new_arr[i])):
            new_arr[i, j, 0] = new_R[i][j]
            new_arr[i, j, 1] = new_G[i][j]
            new_arr[i, j, 2] = new_B[i][j]
    
    new_arr = new_arr.astype('uint8')

    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
def zoomout():
    # Load selected image
    cd = window.filename
    img = PIL.Image.open(str(cd))
    img = img.convert("RGB")
    x, y = img.size
    new_arr = PIL.Image.new("RGB", (int(x / 2), int(y / 2)))
    
#     img_arr = np.asarray(img)
#     new_size = ((img_arr.shape[0])*2, (img_arr.shape[1]*2), img_arr.shape[2])
#     new_arr = np.full(new_size, 255)
    r = [0, 0, 0, 0]
    g = [0, 0, 0, 0]
    b = [0, 0, 0, 0]

    for i in range(0, int(x/2)):
        for j in range(0, int(y/2)):
            r[0], g[0], b[0] = img.getpixel((2 * i, 2 * j))
            r[1], g[1], b[1] = img.getpixel((2 * i + 1, 2 * j))
            r[2], g[2], b[2] = img.getpixel((2 * i, 2 * j + 1))
            r[3], g[3], b[3] = img.getpixel((2 * i + 1, 2 * j + 1))
            new_arr.putpixel((int(i), int(j)), (int((r[0] + r[1] + r[2] + r[3]) / 4), int(
                (g[0] + g[1] + g[2] + g[3]) / 4), int((b[0] + b[1] + b[2] + b[3]) / 4)))
    
    new_arr = np.uint8(new_arr)

    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()

#crop just 200x200 first array of image
def crop():
    
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)

    x1 = int(entryX1.get())
    x2 = int(entryX2.get())
    
    y1 = int(entryY1.get())
    y2 = int(entryY2.get())
    
    #slicing matriks gambar dan dimasukkan ke matriks a
    img = img[y1:y1+y2, x1:x1+x2]
    
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
    
# return true if the image is grayscale, false if not
def is_grey_scale(img_path):
    im = PIL.Image.open(img_path).convert('RGB')
    w, h = im.size
    for i in range(w):
        for j in range(h):
            r, g, b = im.getpixel((i, j))
            if r != g != b:
                return False
    return True


# same rgb histogram of image to static/img
def histogram_rgb():
    # Load selected image
    img_path = window.filename
    img = PIL.Image.open(img_path)
    img_arr = np.asarray(img)
    if is_grey_scale(img_path):
        g = img_arr[:, :, 0].flatten()
        data_g = Counter(g)
        plt.bar(list(data_g.keys()), data_g.values(), color='black')
        plt.savefig(f'static/img/grey_histogram.jpg', dpi=300) #save to folder
        plt.clf()
    else:
        r = img_arr[:, :, 0].flatten()
        g = img_arr[:, :, 1].flatten()
        b = img_arr[:, :, 2].flatten()
        data_r = Counter(r)
        data_g = Counter(g)
        data_b = Counter(b)
        data_rgb = [data_r, data_g, data_b]
        warna = ['red', 'green', 'blue']
        data_hist = list(zip(warna, data_rgb))
        for data in data_hist:
            plt.bar(list(data[1].keys()), data[1].values(), color=f'{data[0]}')
            plt.savefig(f'static/img/{data[0]}_histogram.jpg', dpi=300) #save to folder
            plt.clf()
            
# create convolution for blur,edge detection, sharpening from kernel parameter
def convolution(img, kernel):
    h_img, w_img, _ = img.shape
    out = np.zeros((h_img, w_img), dtype=np.float)
    new_img = np.zeros((h_img, w_img, 3))
    if np.array_equal((img[:, :, 1], img[:, :, 0]), img[:, :, 2]) == True:
        array = img[:, :, 0]
        for h in range(h_img-2):
            for w in range(w_img-2):
                S = np.multiply(array[h:h+3, w:w+3], kernel)
                out[h, w] = np.sum(S)
        out_ = np.clip(out, 0, 255)
        for channel in range(3):
            new_img[:, :, channel] = out_
    else:
        for channel in range(3):
            array = img[:, :, channel]
            for h in range(h_img-2):
                for w in range(w_img-2):
                    S = np.multiply(array[h:h+3, w:w+3], kernel)
                    out[h, w] = np.sum(S)
            out_ = np.clip(out, 0, 255)
            new_img[:, :, channel] = out_
    new_img = np.uint8(new_img)
    return new_img

# edge detection
def edge_detection():
    img_path = window.filename
    img = PIL.Image.open(img_path)
    img_arr = np.asarray(img, dtype=np.int)
    kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])#3x3 kerne
    new_arr = convolution(img_arr, kernel)
    new_arr = np.uint8(new_arr)

    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()

# bluring image
def blur():
    img_path = window.filename
    img = PIL.Image.open(img_path)
    img_arr = np.asarray(img, dtype=np.int)
    kernel = np.array([[1/16, 1/8, 1/16],   #3x3 kernel
                [1/8, 1/4, 1/8],
                [1/16, 1/8, 1/16]])
    new_arr = convolution(img_arr, kernel)
    new_arr = np.uint8(new_arr)

    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    

# sharpening image
def sharpening():
    img_path = window.filename
    img = PIL.Image.open(img_path)
    img_arr = np.asarray(img, dtype=np.int)
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) # 3X3 kernel
    new_arr = convolution(img_arr, kernel)
    new_arr = np.uint8(new_arr)

    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(new_arr))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
def threshold():
    
    img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    img = cv2.resize(img, (852,480), interpolation = cv2.INTER_AREA)
    
    h_img, w_img, channel = img.shape
    
    segmented_image = np.zeros((h_img, w_img, channel))
    
    threshold = [int(R.get()),int(G.get()),int(B.get())]
    
    red_range = (threshold[0] > img[:,:,0])
    green_range = (threshold[1] > img[:,:,1])
    blue_range = (threshold[2] > img[:,:,2])
    
    valid_range = np.logical_and(red_range, green_range, blue_range)

    img[valid_range] = 255
    
    img[np.logical_not(valid_range)] = 0
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
# get neighbors
def get8n(x, y, shape):
    out = []
    maxx = shape[1]-1
    maxy = shape[0]-1

    #top left
    outx = min(max(x-1,0),maxx)
    outy = min(max(y-1,0),maxy)
    out.append((outx,outy))

    #top center
    outx = x
    outy = min(max(y-1,0),maxy)
    out.append((outx,outy))

    #top right
    outx = min(max(x+1,0),maxx)
    outy = min(max(y-1,0),maxy)
    out.append((outx,outy))

    #left
    outx = min(max(x-1,0),maxx)
    outy = y
    out.append((outx,outy))

    #right
    outx = min(max(x+1,0),maxx)
    outy = y
    out.append((outx,outy))

    #bottom left
    outx = min(max(x-1,0),maxx)
    outy = min(max(y+1,0),maxy)
    out.append((outx,outy))

    #bottom center
    outx = x
    outy = min(max(y+1,0),maxy)
    out.append((outx,outy))

    #bottom right
    outx = min(max(x+1,0),maxx)
    outy = min(max(y+1,0),maxy)
    out.append((outx,outy))

    return out

def region_growth():
    list = []
    cv_img = cv2.imread(window.filename)[...,::-1]#OpenCV Use BGR COLOR
    cv_img = cv2.resize(cv_img, (852,480), interpolation = cv2.INTER_AREA)
    cv_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)
    
    segmented_image = np.zeros_like(cv_img)
    
    x,y = int(xStart.get()), int(yStart.get())
    
    threshold = int(thresh.get())
    
    list.append((x,y))
    
    processed = []
    
    while (len(list) > 0):
        pix = list[0]
        
        segmented_image[pix[0], pix[1]] = 255
        
        for coord in get8n(pix[0], pix[1], cv_img.shape):
            if cv_img[coord[0], coord[1]] <= threshold:
                segmented_image[coord[0], coord[1]] = 255
                
                if not coord in processed:
                    list.append(coord)
                
                processed.append(coord)
                
        list.pop(0)
                        
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(cv_img))
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()

# Bolding the image
def dilasi():
    cd = window.filename
    img = PIL.Image.open(str(cd))
    img = img.convert("RGB")
    arr_gray = np.asarray(img)

    h, w, _ = arr_gray.shape
    threshold = 150
    new_array = np.full((h,w),0)
    for i in range(h):
        for j in range(w):
            if arr_gray[i][j][0] > threshold:
                new_array[i][j] = 0
                new_array[i][j] = 0
                new_array[i][j] = 0
            else:
                new_array[i][j] = 255
                new_array[i][j] = 255
                new_array[i][j] = 255
    kernel = [[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255],[255,255,255,255,255,255]]
    array_dilatasi = np.full((h,w),0)
    for i in range(h):
        for j in range(w):
            if (new_array[i,j] == 255) :
                for x in range(len(kernel)):
                    for y in range(len(kernel[1])):
                        if(i+x<h and j+y<w) :
                            r = max(new_array[i+x,j+y],kernel[x][y])
                            array_dilatasi [i+x,j+y] = r
                            
    from PIL import Image
    img_new = Image.fromarray(array_dilatasi)
    img_new = img_new.convert("RGB")
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = img_new)
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
  
'''For thinner image'''
def erosi():
    cd = window.filename
    img = PIL.Image.open(str(cd))
    img = img.convert("RGB")
    arr_gray = np.asarray(img)

    h, w, _ = arr_gray.shape
    threshold = 150
    new_array = np.full((h,w),0)
    for i in range(h):
        for j in range(w):
            if arr_gray[i][j][0] > threshold:
                new_array[i][j] = 0
                new_array[i][j] = 0
                new_array[i][j] = 0
            else:
                new_array[i][j] = 255
                new_array[i][j] = 255
                new_array[i][j] = 255
    kernelerosi = [[255,255,255,255,255,255],[255,255,255,255,255,255]]                      
    array_erosi = np.full((h,w),0)
    for i in range(h):
        for j in range(w):
            if (new_array[i,j] == 255) :
                array_erosi[i,j] = np.min(new_array[i:i+len(kernelerosi),j:j+len(kernelerosi[0])])
    
    from PIL import Image
    img_new = Image.fromarray(array_erosi)
    img_new = img_new.convert("RGB")
    
    # Convert Numpy array to Image
    Image = PIL.ImageTk.PhotoImage(image = img_new)
    
    # Delete image from previous canvas (if available), then add the new image
    C.delete("all")
    C.create_image(0, 0, image = Image, anchor=tkinter.NW)
    window.mainloop()
    
```

```python
def compression():
    global frame
    img = window.filename
    temp = img 
    cv_img = cv2.imread(window.filename)
    cv_img = cv2.resize(cv_img, (852,480), interpolation = cv2.INTER_AREA)
    cv_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB)
    print(cv_img.shape)
    if temp[-4] == '.' : #untuk merapihkan nama untuk ditampilkan ke GUI
        temp = temp[0:-4]
    else:
        temp = temp[0:-5]
    arr = np.zeros_like(cv_img) #array yang seukuran dengan gambar yang ingin dicompress
    # memulai kompresi
    for x in range (cv_img.shape[0]):
        for y in range(cv_img.shape[1]):
            for z in range(cv_img.shape[2]):
                # mengisi arr dengan nilai kompresi menggunakan nilai piksel yang sudah dipetakan pada nilai dictionary
                arr[x,y,z] = dictCompres[cv_img[x,y,z]]
    cv2.imwrite(temp+'_compressed.jpg', arr)
    frame.destroy()
    frame = tk.Frame(bottomFrame0)
    frame.pack()
    info = tk.Label(frame, text="Gambar sudah di compress dan disimpan pada direktori "+ temp+'_compressed.jpg. Untuk membuka file yang sudah di compress, klik tombol Open Compressed Image pada menu bar dibawah')
    info.pack()
    root.mainloop()
    
def decompression():
    #meminta user untuk menginputkan gambar
    dirImg = filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("all files","*.*"), ("jpeg files","*.jpg")))
    cv_img = cv2.imread(dirImg)
    
    #resize gambar untuk mempercepat komputasi
    cv_img = cv2.resize(cv_img, (800,600), interpolation = cv2.INTER_AREA)
    
    #memulai decompressi
    for x in range (cv_img.shape[0]):
        for y in range(cv_img.shape[1]):
            for z in range(cv_img.shape[2]):
                # save image di python menyebabkan nilai kompresi melebihi 51, dimana nilai tersebut diluar rentang dictionary
                # oleh karena itu saya menjadikan nilai kompresi 52 menjadi 51
                if cv_img[x,y,z] > 51:
                    cv_img[x,y,z] = 51
    #memetakan nilai kompresi dengan dictionary Decompressi
    for x in range (cv_img.shape[0]):
        for y in range(cv_img.shape[1]):
            for z in range(cv_img.shape[2]):
                cv_img[x,y,z] = dictDecompres[cv_img[x,y,z]]
    
    cv2.imshow('Open Compressed Image', cv_img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```
